```python
import numpy as np
import matplotlib.pyplot as plt
from keras_preprocessing.image import load_img , img_to_array

%matplotlib inline
```

```python
X_img=load_img("/content/thispersondoesntexist.jpg",target_size=(256,256),color_mode="grayscale"); #loading image and then convert it into grayscale and with it's target size 
X=img_to_array(X_img); #convert image into array
X =np.reshape(X,(256,256))
print(X.shape)
plt.imshow(X_img,cmap='gray')
plt.axis('off')
```

```python
X_meaned = X - np.mean(X , axis = 0)
print (X_meaned)
```

```python
cov_mat = np.cov(X_meaned , rowvar = False)   # rowvar = False >> work with X_meaned as transposed
print(cov_mat.shape)
```

```python
eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)
print(eigen_values.shape)
print(eigen_vectors.shape)
```

```python
#sort the eigenvalues in descending order
sorted_index = np.argsort(eigen_values)[::-1]
print (sorted_index)
sorted_eigenvalue = eigen_values[sorted_index]
#similarly sort the eigenvectors 
sorted_eigenvectors = eigen_vectors[:,sorted_index]
print (sorted_eigenvectors)
```

```python
n_components0 = 10 #you can select any number of components.
eigenvector_subset0 = sorted_eigenvectors[:,0:n_components0] 
n_components1 = 20 #you can select any number of components.
eigenvector_subset1 = sorted_eigenvectors[:,0:n_components1] 
n_components2 = 50 #you can select any number of components.
eigenvector_subset2 = sorted_eigenvectors[:,0:n_components2] 
n_components3 = 100 #you can select any number of components.
eigenvector_subset3 = sorted_eigenvectors[:,0:n_components3] 
```

```python
X_reduced0 = np.dot(eigenvector_subset0.transpose(),X_meaned.transpose()).transpose()
X_reduced1 = np.dot(eigenvector_subset1.transpose(),X_meaned.transpose()).transpose()
X_reduced2 = np.dot(eigenvector_subset2.transpose(),X_meaned.transpose()).transpose()
X_reduced3 = np.dot(eigenvector_subset3.transpose(),X_meaned.transpose()).transpose()

```

```python
X_after = np.dot(X_reduced0,eigenvector_subset0.transpose())
X_after = X_after + np.mean(X , axis = 0)
plt.imshow(X_after,cmap='gray')
plt.axis('off')
```

```python
X_after = np.dot(X_reduced1,eigenvector_subset1.transpose())
X_after = X_after + np.mean(X , axis = 0)
plt.imshow(X_after,cmap='gray')
plt.axis('off')
```

```python
X_after = np.dot(X_reduced2,eigenvector_subset2.transpose())
X_after = X_after + np.mean(X , axis = 0)
plt.imshow(X_after,cmap='gray')
plt.axis('off')
```

```python
X_after = np.dot(X_reduced3,eigenvector_subset3.transpose())
X_after = X_after + np.mean(X , axis = 0)
plt.imshow(X_after,cmap='gray')
plt.axis('off')
```
